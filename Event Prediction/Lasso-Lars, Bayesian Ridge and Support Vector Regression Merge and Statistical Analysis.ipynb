```python
pred_2_df =  pd.read_csv("/home/natasha/Downloads/Lab11/lab11_temp/finalpred_2.tsv",
                  na_values=['-'],
                  sep = '\t'
                )
pred_2_df.head(10)
```

```python
sum_err_br_a =  pred_merged['err_br_a'].sum()
sum_err_br_r =  pred_merged['err_br_r'].sum()
sum_err_br_p =  pred_merged['err_br_p'].sum()
sum_err_br_d =  pred_merged['err_br_d'].sum()
sum_err_br_o =  pred_merged['err_br_o'].sum()
sum_err_br_t =  pred_merged['err_br_t'].sum()
sum_err_br = sum_err_br_a + sum_err_br_r + sum_err_br_p + sum_err_br_d + sum_err_br_o + sum_err_br_t
print sum_err_br
avg_e_br = sum_err_br / (pred_merged['trial_id'].count() * 6)
print avg_e_br
val_e_br = math.sqrt(avg_e_br)
print val_e_br
```
